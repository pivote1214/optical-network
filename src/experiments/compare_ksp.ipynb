```python
import pickle
import networkx as nx

from src.utils.paths import GRAPH_DIR
from src.utils.graph import calc_path_length

with open(GRAPH_DIR / "NSF.pickle", "rb") as f:
    nsf: nx.DiGraph = pickle.load(f)
with open(GRAPH_DIR / "EURO16.pickle", "rb") as f:
    euro16: nx.DiGraph = pickle.load(f)

for u in nsf.nodes:
    for v in nsf.nodes:
        if u < v:
            all_paths = list(nx.all_simple_paths(nsf, u, v))
            hop = sorted(all_paths, key=lambda path: (len(path), calc_path_length(nsf, path)))[:2]
            length = sorted(all_paths, key=lambda path: (calc_path_length(nsf, path), len(path)))[:2]
            if hop[0] != length[0] or hop[1] != length[1]:
                print(u, v)
                print(hop, calc_path_length(nsf, hop[0]), calc_path_length(nsf, hop[1]))
                print(length, calc_path_length(nsf, length[0]), calc_path_length(nsf, length[1]))
```
