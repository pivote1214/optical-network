```python
import os
import sys
sys.path.append('..')

network_dir = "/Users/pivote1214/Documents/Codes/optical-network/data/network"

from utils import graph

name2graph = {
    'N6': graph.N6, 
    'N6S9': graph.N6S9, 
    'NSFNET': graph.NSF, 
    'RING': graph.RING, 
    'EURO16': graph.EURO16, 
    'JPN12': graph.JPN12, 
    'JPN25': graph.JPN25, 
    'US24': graph.US24
    }

for name, graph in name2graph.items():
    # タプルの1つ目，2つ目の順に昇順ソート
    graph = sorted(graph, key=lambda x: (x[0], x[1]))
    # csvファイルに書き込み
    columns = ['source', 'target', 'weight']
    with open(os.path.join(network_dir, f"{name}.csv"), 'w') as f:
        f.write(','.join(columns) + '\n')
        for edge in graph:
            f.write(','.join(map(str, edge)) + '\n')
    print(f"write {name}.csv")
    nodes = {node for edge in graph for node in edge[:2]}
    print(f"nodes: {len(nodes)}")
    print(f"edges: {len(graph)}")
```
