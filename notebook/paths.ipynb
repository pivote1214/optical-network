```python
import os
import sys
import pickle
sys.path.append('..')

import networkx as nx
import matplotlib.pyplot as plt

from utils.namespaces import PATHS_DIR
from utils.network import load_network

network_name = 'GRID3'
algorithm = 'k-shortest-paths'
path_weights = ['hop', 'physical-distance', 'expected-used-slots']
n_paths = 3

path_weight = 'hop'
paths_file = os.path.join(PATHS_DIR, algorithm, network_name, f'path-weight_{path_weight}', f'n-paths_{n_paths}.pkl')
paths = pickle.load(open(paths_file, 'rb'))
# パスを可視化
network = load_network(network_name)
source, target = 0, 8
# network描画
pos = nx.spring_layout(network)
nx.draw(network, pos, with_labels=True, node_size=300, node_color='w', font_size=8, font_color='k')
# source -> targetの経路を描画
paths1 = paths[(source, target)]
print(paths1)
for i, path in enumerate(paths1):
    path_edges = list(zip(path[:-1], path[1:]))
    if i == 0:
        nx.draw_networkx_edges(network, pos, edgelist=path_edges, edge_color='r', width=2, alpha=0.5)
    elif i == 1:
        nx.draw_networkx_edges(network, pos, edgelist=path_edges, edge_color='g', width=2, alpha=0.5)
    elif i == 2:
        nx.draw_networkx_edges(network, pos, edgelist=path_edges, edge_color='b', width=2, alpha=0.5)
```
